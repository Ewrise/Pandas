```python
import pandas as pd
```

```python
df = pd.read_table("veri.txt")
df

# tablar ile ayrılmış veri setimizi içe aktardık.
```

```python
df = pd.read_table("veri2.txt", sep= ",")
df

# virgüllerle ayrılmış veri setimizi de aynı methodla içe aktarabiliriz.
# ancak bunun için sep parametresi kullanmamız gerekir.
# veri setinde sütun isimleri de verideki ilk satır olarak alındı.
# bu istemediğimiz bir şey
```

```python
df = pd.read_table("veri2.txt", sep= ",", header= None)
df

# header= parametresiyle sütun ismimizin veri setinde olmadığını
# belirtirsek ilk satırı sütun isimleri olarak almaz.
# indeks numaralarıyla belirtir.
```

```python
df = pd.read_table("veri2.txt", sep= ",", header= None, names= ["isim", "puan", "cinsiyet"])
df

# names= parametresiyle sütun isimlerini belirleyebiliriz.
```

```python
df = pd.read_table("veri2.txt", sep= ",", header= None, 
    names= ["isim", "puan", "cinsiyet"], index_col= "isim")
df

# index_col= parametresiyle bir sütunu satır olarak alabiliriz.
```

```python
df = pd.read_table("veri3.txt", sep= ",")
df

# veri3.txt dosyasını içe aktardık.
```

```python
df = pd.read_table("veri3.txt", sep= ",", index_col= ["ders", "isim"])
df

# Satırları ders ve isim sütunlarıyla değiştirdik.
```

```python
df = pd.read_table("veri4.txt", sep= ",")
df

# Direkt okuttuğumuzda yorum satırlarını da okuyor.
```

```python
df = pd.read_table("veri4.txt", sep= ",", skiprows= [0, 2])
df

# Okutmadan geçmek istediğimiz satırları skiprows= parametresiyle belirleyebiliriz.
```

```python
df = pd.read_table("veri4.txt", sep= ",", skiprows= [0, 2], usecols= [0, 1])
df

# usecols= parametresiyle de sadece istediğimiz satırları okutabiliriz.
```

```python
df = pd.read_table("veri5.txt", sep= ",")
df

# Eksik verili veri setimizi okuttuğumuzda veri olmayan yerlere NaN değeri gelir.
```

```python
pd.isnull(df)

# isnull() methoduyla eksik veri olup olmadığını sorgularız.
# True eksik veri olduğunu False olmadığını belirtir.
```

```python
df = pd.read_table("veri5.txt", sep= ",", na_values= ["n.a", -5])
df

# na_values= parametresiyle normalde eksik veya yanlış olan verileri
# pythonun eksik veri olarak algılamasına yardımcı olabiliriz.
```

```python
eksik_veri= {"puan": -5, "cinsiyet": ["n.a", "Efe"]}
df = pd.read_table("veri5.txt", sep= ",", na_values= eksik_veri)
df

# eksik verileri bir sözlük yapısına atayarakta bu işlemi yapabiliriz.
# Burada bunu kullanmamızdaki sebep isim verisinde de Efe verisinin bulunması
# Eğer na_values= parametresine direkt Efe yazsaydık isim sütunundaki
# Doğru veri de NaN değeriyle değişecekti.
```

```python
import sys

df.to_csv(sys.stdout, sep= "|", na_rep= "NULL")

# eksik verilerin NaN değilde NULL olarak görünmesini istedik.
```

```python
import sys

df.to_csv(sys.stdout, sep= "|", index= False, header= False)

# Satır ve sütun isimlerini görmek istemezsek
# index= ve header= parametrelerine False girebiliriz.
```
