```python
import pandas as pd
```

```python
veri = {"isim": ["Ali", "Can", "Efe", "Ayşe", "Buse", "Alp", "Nur"], 
"puan": [90, 80, 85, 75, 95, 60, 65],
"spor": ["Güreş", "Futbol", "Kayak", "Yüzme", "Tenis", "Karate", "Sörf"],
"cinsiyet": ["E", "E", "E", "K", "K", "E", "K"]}

df = pd.DataFrame(veri)
df

# İlk dataframe 'imizi oluşturduk.
# Series 'ten farklı olarak birden fazla satırdan oluştuğunu görebiliyorsunuz.
```

```python
df.head()

# İlk 5 veriye head() metoduyla ulaşabiliriz.
```

```python
df.tail(2)

# Eğer parametre olarak bir sayı girersek girdiğimiz sayı kadar bize sondan veri yazdırır.
```

```python
df = pd.DataFrame(veri, columns= ["isim", "spor", "cinsiyet", "puan"])
df

# Sütunların sıralamasını kendimizde columns= parametresiyle belirtebiliriz.
```

```python
df = pd.DataFrame(veri, columns= ["isim", "spor", "cinsiyet", "puan", "yaş"])
df

# Bu şekilde sütun eklemesi de yapabiliriz. 
# Ancak yaş sütununa bir veri girmediğimiz için veriler NaN yani eksik veri olarak gözükecektir.
```

```python
df = pd.DataFrame(veri, columns= ["isim", "spor", "cinsiyet", "puan", "yaş"],
    index= ["bir", "iki", "üç", "dört", "beş", "altı", "yedi"])
df

# index numaralarını da index= parametresiyle kendimiz isimlendirebiliriz.
```

```python
df.loc[["bir"]]

# İstediğimiz satıra ulaşmak için loc özelliğini kullanıyoruz.
# Burada iç içe iki köşeli parantez kullandığımıza dikkat edin.
```

```python
df.loc[["bir", "yedi"]]

# Satırlarda birden fazla satır yazdırmak istediğimizde sadece virgülle ayırmamız yeterli oluyor.
```

```python
df["yaş"] = 18
df

# yaş sütununa 18 değerini atamış olduk.
```

```python
değer = [17, 18, 19, 17, 18, 20, 18]
df["yaş"] = değer
df

# Eğer dataframe içine bir liste atayacak olursak bunları bir tanıma atamamız gerekiyor.
# Girilen değerlerle satır sayısınında eşit olması gerekiyor yoksa hata alırız.
# Örneğin burada 7 satırlı bir dataframe 'imiz var bu yüzden 7 değerli bir liste tanımlamamız gerekiyordu.
```

```python
df["geçti"] = df.puan > 70
df

# geçti adında sütun oluşturduk değerlerine de 70ten yüksek alanları True olarak yazdırdık.
```

```python
del df["geçti"]
df

# del özelliği ile de dataframe 'de sütun silme işlemi yapabiliyoruz.
```

```python
notlar = {"Mat": {"Ali": 85, "Efe": 90, "Nur": 96}, "Fiz": {"Ali": 90, "Efe": 80, "Nur": 75}}

puan = pd.DataFrame(notlar)
puan

# İç içe sözlük yapısını da dataframe 'e çevirebilmeniz mümkün.
# Dıştaki sözlüğün keylerini sütun olarak
# İçteki sözlüğün keylerini satır olarak alır.
# value değerleri kesişimlerine yazılır.
```

```python
puan.T

# T ile dataframe 'in transpozu alınabilir.
# Yani satır ve sütunların yeri değişmiş olur.
```

```python
puan.index.name = "isim"
puan.columns.name = "ders"
puan

# Satır ve sütunlara da isim verme işlemi bu şekilde yapılıyor.
# Tercihen kullanılabilir. 
# Ancak çok kullanışlı olduğunu söyleyemem.
```

```python
puan.values

# values özelliği dataframe 'in değerlerini iki boyutlu ndarray 'e dönüştürür.
```
