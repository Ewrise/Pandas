```python
import pandas as pd
import numpy as np
```

```python
s = pd.Series(range(5), index= ["e", "d", "a", "b", "c"])
s

# s değişkenine Series veri tipi oluşturduk
```

```python
s.sort_index()

# s değişkenini index'ine göre sıraladık. 
# Yani satır isimlerine göre.
```

```python
df = pd.DataFrame(np.arange(12).reshape(3, 4), index= ["iki", "bir", "üç"], columns= ["d", "a", "b", "c"])
df

# df değişkeninde dataframe oluşturduk.
# satır ve sütun isimlerini belirledik.
```

```python
df.sort_index()

# sort_index() metoduyla veri setini satır isimlerine göre sıraladık.
```

```python
df.sort_index(axis= 1)

# axis= 1 parametresiyle sütunlarda sıralama yapabiliriz.
```

```python
df.sort_index(axis= 1, ascending= False)

# ascending= False parametresiyle sıralamayı tersten yapabiliriz.
```

```python
s2 = pd.Series([5, 3, -1, 9])
s2

# s2 değişkeninde Series veri yapısı oluşturduk.
```

```python
s2.sort_values()

# sort_values() methoduyla verilere göre sıralama yapabiliyoruz.
```

```python
s2 = pd.Series([5, np.nan, 3, -1, 9])
s2.sort_values()

# Verilere göre sıralama yapıldığında eğer eksik veri varsa o satır sona koyulur.
```

```python
df2 = pd.DataFrame({"a": [5, 3, -1, 9], "b": [1, -2, 0, 5]})
df2

# df2 değişkeninde Dataframe tipinde veri oluşturduk.
```

```python
df2.sort_values(by= "b")

# by= "b" parametresiyle b sütununun verilerine göre sıralama yapmasını söyledik.
# Bütun satırlar da buna göre sıralandı.
# a sütununda yanlış bir sıralanma olduğunu görebilirsiniz.
```

```python
df2.sort_values(by= ["b", "a"])

# İki tane sıralama yapmasını söylediğimizde de istediğimiz sonucu alamıyoruz.
# Çünkü bunda aynı değere sahip satırlar bulunmuyor.
# Büyük verilerde bunu tekrar deneyip farkı göreceğiz.
```

```python
veri = pd.read_csv("vgsales.csv")
veri

# vgsales.csv dosyasını içe aktardık ve yazdırdık.
```

```python
veri.head()

# dataframe'imizin ilk 5 satırını yazdırdık.
```

```python
veri["Name"].sort_values()

# dataframe'imizdeki Name sütununu aldık ve ondaki verileri sıraladık.
```

```python
veri.Name.sort_values()

# Sütun alma işlemini bu şekilde de belirtebiliriz.
```

```python
type(veri.Name.sort_values())

# Tek sütun aldığımız için bu artık Series veri tipine dönüşmüştür.
```

```python
veri.Name.sort_values(ascending= False)

# ascending= False parametresiyle yine ters sıralama yapabiliriz.
```

```python
veri.sort_values("Name")

# Bütün veri setini tek bir satıra göre de sıralayabiliriz.
# Bunun için gördüğünüz gibi sort_values() metodunun parametresi olarak istediğimiz sütun adını girebiliriz.
```

```python
veri.sort_values("Year")

# Bu seferde yıllara göre sıralama yaptık.
```

```python
veri.sort_values(["Year", "Name"])

# İkili sıralama yapmak için liste içine aldık.
# Önce year sütununa göre sonra name sütununa göre sıralamasını istedik.
# Eğer year sütununda aynı veriye sahip satırlar varsa bu sefer name sütununa göre sıralamaya kendi aralarında devam edecektir.
```
