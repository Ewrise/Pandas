```python
def check_df(df):
  print("**********head********************")
  print(df.head())
  print("**********tail********************")
  print(df.tail())
  print("**********info********************")
  print(df.info())
  print("**********columns*****************")
  print(df.columns)
  print("**********null_values*************")
  print(df.isnull().values.any())  
  print("**********shape*******************")
  print(df.shape)
  print("**********describe****************")
  print(df.describe().T)
  print("**********null_sum****************")
  print(df.isnull().sum())
  print("***********Corr*******************")
  print(df.corr())
  print("***********dtypes*****************")
  print(df.dtypes)
 
check_df(df)
```
