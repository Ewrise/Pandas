```python
import pandas as pd
```

```python
review = pd.read_csv("ReviewContent.csv")

print(review)

# Cvs dosyamızı okutuyoruz.
```

```python
print(review["City"])

# Dataframe 'imizdeki sadece City sütunundaki verileri yazdırdık.
```

```python
print(review["City"] == "London")

# Yalnızca City sütunundaki London olan verileri içlerinden aldık.
# True değerler görülenler London değeri olan verilerdir.
```

```python
print(review[review["City"] == "London"])

# İkinci bir review kullandık.
# City sütundaki London verisi olan satırları bize yazdırdı.
# Satır sayısına da baktığımızda eleme yaptığını görebiliriz.
# Normalde 1000 adet satırımız varken yaptığımız bu işlemle satır sayısını 104 satıra indirdik.
# Bu da demek oluyor ki London verisi olan 104 verimiz bulunuyormuş.
```

```python
mask = (review["City"] == "London")

print(review[mask])

# Daha düzenli görünmesi için aynı işlemi bu şekilde yapmak daha anlaşılır kodlar yazmanızı sağlayabilir.
```

```python
mask = review["Room Type"] == "Private room"

print(review[mask])

# Daha anlaşılır olması adına benzer bir örnek yaptık.
# Room Type sütunlarını inceledik.
# Bu sütunda Private room olan verileri ekrana yazdırdık.
# Toplam 304 adet satıra filtrelemiş olduk.
```

```python
mask = review["Room Type"] != "Private room"

print(review[mask])

# Bu sefer olanı değil olmayanı aradık.
# != sembolü bildiğiniz gibi eşit olmayan demektir. 
# mask tanımımıza True değer dönmesi için Room sütununda Private room yazmaması gerekiyor.
# Yani Private room olmayan odaları yazdırmış olduk.
```

```python
mask1 = review["Room Price"] > 60 
mask2 = review["Room Price"] < 100

mask = mask1 & mask2

print(review[mask])

# Bu sefer sınırlarımızı daha da geliştirdik ve iki filtreleme kullandık.
# Birincisinde Room Price değerinin 60 tan büyük olmasını
# İkincisinde ise 100 den küçük olmasını istedik.
```

```python
mask = (review["Room Price"] > 60) & (review["Room Price"] < 100)

print(review[mask])

# Bunları daha basit bir şekilde tek bir ifade içine de sığdırabiliriz.
# Burada & işareti and anlamında kullanılıyor.
# Yani her iki şartında True değeri vermesi gerekiyor.
```

```python
# | -> işaretimizde or anlamına gelir.
# Tek bir şartın True değer vermesi yeterlidir.
```

```python
mask = (review["Room Price"] > 200) | (review["Room Availability"] > 300)

print(review[mask])

# Oldukça zor bir şartlar koymamıza rağmen or ifadesi verdiğimiz için iki şarttan birinin
# True olması yazdırılması için yeterli hale getirdi.
```

```python
# isin() metodu

mask = (review["City"] == "Madrid") | (review["City"] == "London") | (review["City"] == "Amsterdam")

print(review[mask])

# Farklı filtreler uygulayarak yazdırma işlemini yaptık.
```

```python
mask = (review["City"] == "Madrid") | (review["City"] == "London") | (review["City"] == "Amsterdam") | (review["City"] == "Paris")

print(review[mask])

# Paris verisini de ekledik. 
# Ancak her veri eklediğimizde böyle uzatarak devam etmeyeceğiz. 
# Bunun için isin() metoduna ihtiyacımız var.
```

```python
mask = review["City"].isin(["Madrid", "London", "Amsterdam", "Paris"])

print(review[mask])

# Yukarıda yazdığımız uzun uzadıya kodları çok daha basit bir şekilde isin() metodu kullanarak yazdırdık.
# Hiçbir farkı olmadığı kendiniz de kontrol edebilirsiniz.
# Ancak bu | yani or yerine kullanılan bir metodtur.
```

```python
# between() metodu

mask = review["Room Price"].between(80, 100)

print(review[mask])

# between ingilizceden de bildiğiniz gibi 2 parametre arasındaki bütün verileri bize yazdırıyor.
# Burada dikkat etmeniz gereken 80 ve 100 'ünde bu yazdırılanlara dahil olmasıdır.
```

```python
mask = review["Review Date"].between("2013-01-01", "2014-01-01")

print(review[mask])

# 2013ten 2014 e kadar olan bütün verileri yazdırdık.
# Bunu da başka bir örnek olarak inceleyebilirsiniz.
```

```python
# duplicated() metodu

review.sort_values("City", inplace= True)

print(review["City"])

# Dataframe 'imizi City sütununa göre sıralamayı yaptık.
# Daha önce gördüğümüz üzere de inplace parametresiyle orjinal dataframe 'imizde değişikliği gerçekleştirdik.
```

```python
print(review["City"].duplicated())

# Bir ifade ile ilk karşılaştığında False değeri yazdırıyor.
# Ancak ondan sonra o ifadeyle ne zaman karşılaşsa bize True ifadesini yazdırıyor.
# Bu duplicated metodunun bize sağladığı bir filtreleme yöntemidir.
```

```python
mask = review["City"].duplicated() == False

print(review[mask])

# maskımızı da kullanarak tekrar eden şehirleri tekrar yazdırmasını engellemiş olduk.
```

```python
# drop_duplicates() metodu

review = pd.read_csv("ReviewContent.csv")

print(review)

# İlk orjinal csv verilerimizi yazdırdık.
```

```python
print(len(review))

# Dataframe 'imizin uzunluğuna baktık. 1000 satırlı olduğunu zaten biliyorduk.
```

```python
print(len(review.drop_duplicates()))

# Yine aynı sonucu aldık. 
# Çünkü bu metodta bir satırdaki bütün veriler aynı olursa aynı olan verilerden birini siler.
# Ancak böyle bir satırımız olmadığı için aynı çıktıyı aldık.
# Değişiklik yaparak deneyelim.
```

```python
print(len(review.drop_duplicates(subset= "City")))

# Bu sefer metodumuza subset= metodunu verdik.
# Bu parametreyle metodumuz bütün satıra değil satırda bizim belirttiğimiz sütunlarda karşılaştırma yapar.
```

```python
print(len(review.drop_duplicates(subset= "Room Type")))

# Bu da farklı bir örnek.
```

```python
# unique(), nunique() metodları

print(review["City"])

# Normal bir sütun filtrelemesi yaptık.
```

```python
print(review["City"].unique())

# unique metodu ingilizceden de bildiğiniz gibi sadece eşsiz olanları bize yazdırır.
# Yani kullanılan bütün şehir isimlerini bu şekilde öğrenebiliriz.
```

```python
print(type(review["City"].unique()))

# Ancak bize verdiği veri numpy bir veridir. Bunu da bilmekte fayda var.
```

```python
print(len(review["City"].unique()))

# 27 tane farklı şehir olduğunu bu metotla da öğrenebiliriz.
```

```python
print(review["City"].nunique())

# nunique metodu ise bize şehir isimlerini yazdırmaz.
# Direkt olarak kaç tane farklı veri olduğunu bize söyler.
# Yani bir nevi len metodununda otomatik olarak kullanıldığını bu metotta söyleyebiliriz.
```
